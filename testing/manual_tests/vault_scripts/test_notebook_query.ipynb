# Query with Session

```python
def query_session(code_text = None, description_embedding = None, description_text = None):
    print("QUERY EMBEDDING")
    result = vault.query_embedding_collection(
            embedding_query,
            description_embedding=description_embedding,
            description_text=description_text,
            code_text=code_text,
            filtered=filtered_list,
            use_approx=False,
        )
    print(result)
    print("QUERY RECORD")
    result =  vault.query_record_collection(
        record_text,
        description_embedding=description_embedding,
        description_text=description_text,
        code_text=code_text,
        filtered=filtered_list,
    )
    print(result)
    print("QUERY DOCUMENT")
    result = vault.query_document_collection(
        document_text,
        description_embedding=description_embedding,
        description_text=description_text,
        code_text=code_text,
        filtered=filtered_list,
    )
    print(result)
    print("QUERY FILE")
    result = vault.query_file_collection(
        description_embedding=description_embedding,
        description_text=description_text,
        code_text=code_text,
        filtered=filtered_list,
    )
    print(result)

```
